```python
import torch
import torch.nn as nn

from gsl import GSL
```

```python
class GCLayer(nn.Module):
    """
    Graph convolution layer.
    """
    def __init__(self, in_dim, out_dim):
        super().__init__()
        self.dense = nn.Linear(in_dim, out_dim)

    def forward(self, adj, X):
        adj = adj + torch.eye(adj.size(0)).to(adj.device)
        h = self.dense(X)
        norm = adj.sum(1)**(-1/2)
        h = norm[None, :] * adj * norm[:, None] @ h
        return h
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class AnomalyClassifier(nn.Module):
    def __init__(self, pretrained_ts_model, hidden_gcn=32, hidden_mlp=16):

        super().__init__()
        self.gsl = pretrained_ts_model.gsl[0]
        self.z   = pretrained_ts_model.z
        
        for p in self.gsl.parameters():
            p.requires_grad = False
        self.gsl.eval()
        
        self.gcn = GCLayer(in_dim=1, out_dim=hidden_gcn)
        
        self.pool = nn.AdaptiveMaxPool1d(1)
        
        self.classifier = nn.Sequential(
            nn.Linear(hidden_gcn, hidden_gcn//2),
            nn.ReLU(inplace=True),
            nn.Linear(hidden_gcn//2, hidden_mlp),
            nn.ReLU(inplace=True),
            nn.Linear(hidden_mlp, 2)
        )
    
    def forward(self, x):
        B, N = x.shape
        assert N == self.z.size(0), "Ожидалось 52 признака"
        
        # 1) строим adj из обученного gsl
        idx = torch.arange(N, device=x.device)
        # self.gsl(idx) возвращает (N,N), умножаем на z — маску главной диагонали
        adj = self.gsl(idx) * self.z      # (N,N)
        
        # 2) прогоняем каждый пример через GCLayer
        feats = []
        for i in range(B):
            xi = x[i].unsqueeze(1)        # (N,1)
            hi = self.gcn(adj, xi)        # (N, hidden_gcn)
            # переходим к форме (1, hidden_gcn, N) для пуллинга
            hi = hi.transpose(0,1).unsqueeze(0)
            # Max‑pool по N → (1, hidden_gcn, 1) → squeeze → (1, hidden_gcn)
            pi = self.pool(hi).squeeze(-1)  
            feats.append(pi)
        
        # (B, hidden_gcn)
        feats = torch.cat(feats, dim=0)
        
        # 3) логиты (B,2)
        logits = self.classifier(feats)
        return logits
```

```python
import torch
from torch.utils.data import TensorDataset, DataLoader

# Преобразуем NumPy в тензоры
X_tensor = torch.tensor(X_train, dtype=torch.float32)
y_tensor = torch.tensor(y_train, dtype=torch.long)  # long для CrossEntropyLoss

# Dataset
train_dataset = TensorDataset(X_tensor, y_tensor)

# DataLoader
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
```

```python
device = 'cuda:0'
classifier = AnomalyClassifier(model, hidden_gcn=64, hidden_mlp=32).to('cuda:0')

# оптимизатор и лосс
optimizer = torch.optim.Adam(filter(lambda p: p.requires_grad, classifier.parameters()), lr=1e-4)
criterion = nn.CrossEntropyLoss()

# цикл обучения
for epoch in range(100):
    classifier.train()
    total_loss = 0.0
    for x_batch, y_batch in train_loader:   # x: (B,52), y: (B,)∈{0,1}
        x_batch, y_batch = x_batch.to(device), y_batch.to(device)
        optimizer.zero_grad()
        logits = classifier(x_batch)         # (B,2)
        loss = criterion(logits, y_batch)
        loss.backward()
        optimizer.step()
        total_loss += loss.item() * x_batch.size(0)
    print(f"Epoch {epoch}: train loss = {total_loss/len(train_dataset):.4f}")

classifier.eval()
```

```python
X_val_tensor = torch.tensor(X_test, dtype=torch.float32)
y_val_tensor = torch.tensor(y_test, dtype=torch.long)

val_dataset = TensorDataset(X_val_tensor, y_val_tensor)
val_loader = DataLoader(val_dataset, batch_size=64, shuffle=False)
```

```python
predict_arr = []
true_arr = []
with torch.no_grad():
    correct = 0
    for x_batch, y_batch in val_loader:
        x_batch, y_batch = x_batch.to(device), y_batch.to(device)
        preds = classifier(x_batch).argmax(dim=1)
        correct += (preds == y_batch).sum().item()
        true_arr.append(y_batch.cpu().numpy())
        predict_arr.append(preds.cpu().numpy())
    print("Val Acc:", correct / len(val_dataset))
```
